```python
import pandas as pd
import os.path as path
from sklearn import datasets
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = datasets.load_iris()
iris_path = path.join(path.dirname(datasets.__file__), 'data', 'iris.csv')
```

```python
# Part 1
# Visualization 1: Bar plot of Sepal Width
plt.figure(figsize=(8, 6))
avg_sepal_width = [iris_df[iris_df['species'] == species]['sepal width (cm)'].mean() for species in iris.target_names]
plt.bar(iris.target_names, avg_sepal_width, color='lavender')
plt.title('Average Sepal Width')
plt.ylabel('Sepal Width (cm)')
plt.grid(axis='y')

for bar in plt.gca().patches:
    bar.set_edgecolor('mediumpurple')
    bar.set_linewidth(1.5)

plt.show()
```

```python
#Part 1
# Visualization 2: Scatter plot of Sepal Length vs Sepal Width
plt.figure(figsize=(8, 6))
for species in iris.target_names:
    subset = iris_df[iris_df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], label=species)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```

```python
#Part 1
# Visualization 3: Violin plot of Petal Length

plt.figure(figsize=(8, 6))
parts = plt.violinplot([iris_df[iris_df['species'] == species]['petal length (cm)'] for species in iris.target_names], showmeans=False, showmedians=True)

plt.xticks([1, 2, 3], iris.target_names)
plt.title('Petal Length')
plt.ylabel('Petal Length (cm)')
plt.grid(True)

# Customize the violin plot to be pretty
for pc in parts['bodies']:
    pc.set_facecolor('cornflowerblue')
    pc.set_edgecolor('black')
    pc.set_alpha(0.7)

plt.show()
```

```python
#Part 2
import pandas as pd
import os.path as path
import numpy as np

# Load the Loan Dataset
loan = pd.read_csv(r"C:\Users\abbis\Downloads\archive (7)\LoanDataset - LoansDatasest.csv")
loan_path = path.join(path.dirname(datasets.__file__), 'data', 'loan.csv')

print(loan.head())
```

```python
#Part 2
#Visualization 1: Bar plot of Customer Age
plt.figure(figsize=(8, 6))
age_bins = [18, 30, 40, 50, 60, 70, 80]
age_labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+']
loan['age_bin'] = pd.cut(loan['customer_age'], bins=age_bins, labels=age_labels)
age_bin_counts = loan['age_bin'].value_counts().sort_index()
for i, count in enumerate(age_bin_counts):
    plt.text(i, count + 100, str(count), ha='center', va='bottom')
plt.bar(age_labels, age_bin_counts, color='skyblue')
plt.title('Customer Age Distribution')
plt.ylabel('Number of Customers')
plt.grid(axis='y')
plt.show()
```

```python
#Part 2
# Visualization 2: Scatterplot Age vs Loan Intent
plt.figure(figsize=(8, 6))
plt.xlim(18, 85)  # Shrink the x-axis to range from 18 to 85
for intent in loan['loan_intent'].unique():
    subset = loan[loan['loan_intent'] == intent]
    plt.scatter(subset['customer_age'], subset['loan_intent'], label=intent)
plt.title('Age vs Loan Intent')
plt.xlabel('Age')
plt.ylabel('Loan Intent')
plt.yticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#Part 2
# Visualization 3: Histogram of Loan Terms

plt.figure(figsize=(8, 6))
plt.hist(loan['term_years'], bins=20, color='lightcoral')
plt.title('Loan Term Distribution')
plt.xlabel('Term (years)')
plt.ylabel('Number of Loans')
plt.grid(axis='y')
plt.show()
```
